```python
# pandas 라이브러리 불러오기
import pandas as pd
```

```python
# pandas.read_csv( "파일경로" )
# 깃허브의 www.github.com/hyowakn -> python_lecture 클릭 dataset ->
# kopo_customerdata.csv
csFilePath = "https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv"
customerData = pd.read_csv(csFilePath)
```

```python
# customerData의 TOTAL_AMOUNT 컬럼의 값들을 가져와서 list 형태로 변환
customerTotalList = customerData["TOTAL_AMOUNT"].tolist()
len(customerTotalList)
```

```python
# Step 1: 동일한 값을 반복적으로 제거하기 위해서 정렬 처리를 한다
customerTotalList.sort()
```

```python
# Step 2: customerTotalList의 최소값을 구한다
minValue = min(customerTotalList)
minValue
```

```python
# Step 3: customerTotalList의 최소값인 0의 개수를 구한다.
cntMinValue = customerTotalList.count(minValue)
cntMinValue
```

```python
# Step 4: customerTotalList의 256번째값을 확인한다.
customerTotalList[255]
```

```python
# Step 5: customerTotalList에서 0을 제거한 List를 가져온다
customerMinRvList = customerTotalList[256:]
```

```python
# Step 6: customerMinRvList의 최대값을 구한다
maxValue = max(customerMinRvList)
maxValue
```

```python
# Step 7: customerMinRvList의 최대값인 3103500의 개수를 구한다.
cntMaxValue = customerMinRvList.count(maxValue)
cntMaxValue
```

```python
# Step8: customerMinRvList의 총 길이를 구한다
len(customerMinRvList)
```

```python
# Step 8: customerMinRvList의 마지막 값을 확인한다.
customerMinRvList[299743]
```

```python
# Step 9: customerTotalList에서 3103500을 제거한 List를 가져온다
customerFinalList = customerMinRvList[:299743]
```

```python
# Step 10: customerTotalList에서 평균을 구한다.
sumTotalList = sum(customerFinalList)
lenTotalList = len(customerFinalList)
avgTotalList = sumTotalList / lenTotalList
avgTotalList
```

```python
# Step 11: 계산 값은 정수여야 하므로 int형으로 형변환을 해준다
answer = int(avgTotalList)
answer
```
