```python
# 필요 라이브러리 설치
!pip install bing-image-downloader==1.1.2

import os
os.environ['PYTHONIOENCODING'] = 'utf-8'
import shutil
import random
import bing_image_downloader # bing에서 검색한 결과에 나오는 이미지들을 다운로드하기 위한 라이브러리
from bing_image_downloader import downloader
from datetime import datetime
import numpy as np
import os
import cv2

def collectImg(queryList, imageCnt, trainImgList, trainLabelList):
    try:
        # 현재 날짜와 시간 가져와서 currentDate에 대입
        currentDate = datetime.now()

        dateKey = currentDate.strftime("%Y%m%d")

        directory_path = f"/content/downimg_{dateKey}"
        timeoutStd = 10
        tempList = []
        for i in range(0, len(queryList)):
            downloader.download(queryList[i],
                                limit=imageCnt,
                                output_dir=directory_path,
                                adult_filter_off=True,
                                force_replace=False,
                                timeout=timeoutStd)
        for i in range(0, len(queryList)):
            categoryPath = os.listdir(directory_path)[i]
            print(categoryPath)
            catPath = os.path.join( directory_path, categoryPath )
            print(catPath)
            for j in range(0, imageCnt):
                imagesPath = os.listdir(catPath)[j]
                imgeFullPath = os.path.join( directory_path, categoryPath, imagesPath )
                inImg = cv2.imread( imgeFullPath, cv2.IMREAD_GRAYSCALE )
                IMG_SIZE = 200
                resizedImg = cv2.resize( inImg, (IMG_SIZE, IMG_SIZE) )
                trainImgList.append( resizedImg )
                if categoryPath not in trainLabelList:
                    trainLabelList.append(categoryPath)
        print(f"저장된 라벨 개수: {len(trainLabelList)}")
        print(f"저장된 이미지 개수: {len(trainImgList)}")


    except Exception as e:
        print(str(e))
```

```python
trainImgList = []
trainLabelList = []
queryList = ["행복한 표정", "화난 표정"] # 검색할 이미지 목록
imageCnt = 30 # 가져올 이미지 개수
collectImg(queryList, imageCnt)
```
