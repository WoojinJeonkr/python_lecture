```python
# Step1: 필요 라이브러리 설치
!pip install requests
!pip install bs4

# Step2: 설치한 라이브러리 가져오기
import requests, bs4
import pandas as pd

# Step3: https://wikidocs.net/
targetUrl = "https://wikidocs.net/"


# Step4: targetUrl을 바탕으로 웹 페이지 요청
resp = requests.get(url = targetUrl)

# Step6: 한글로 인코딩
resp.encode = "utf-8"

# Step7: 원 소스 전부 스크랩
htmlObj = resp.text
htmlObj

bsObj = bs4.BeautifulSoup(htmlObj, "html.parser")
bsObj

spanDiv = bsObj.find(name="span", attrs={"class":"text-muted small"})
spanDiv

aLinkTags = spanDiv.findAll(name="a")

# Step13: [aLinkTags[i]의 속성값, aLinkTags[i]의 href 속성 정보의 형태로 넣기 위한
#         totalList 리스트를 빈 리스트로 생성
totalList = []

# Step14: aLinkTags의 길이만큼 반복하면서 totalList에
#         [aLinkTags[i]의 속성값, aLinkTags[i]의 href 속성 정보 넣기
for i in range(0, len(aLinkTags)):
    aLinkTagsText = aLinkTags[i].text
    aLinkTagsHref = aLinkTags[i].attrs["href"]
    if aLinkTagsHref == "/":
        aLinkTagsHref = "https://wikidocs.net/"
    totalList.append([aLinkTagsText, aLinkTagsHref])

# Step15: totalList에 잘 담겼는지 확인
# print(totalList)

bottomMenu = pd.DataFrame(totalList, columns=["페이지 명", "페이지 링크"])

# Step17. bottomMenu 확인
bottomMenu
```
